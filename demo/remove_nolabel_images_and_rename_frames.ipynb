```python
import os
from glob import glob
```

```python
source_image_dir = "../resources/FLL_VAL/images/"
source_label_dir = "../resources/FLL_VAL/labels/"
```

```python
len(os.listdir(source_image_dir)), len(os.listdir(source_label_dir))
```

## 1. Remove no label images

```python
os.listdir(source_label_dir)[:3]
```

```python
cnt = 0
for image_file in os.listdir(source_image_dir):
    label_file = source_label_dir+os.path.splitext(image_file)[0]+".txt"
    cnt += os.path.exists(label_file)
    if not os.path.exists(label_file):
        print("label file does not exit")
        os.remove(os.path.join(source_image_dir,image_file))
print(cnt)
```

```python
len(os.listdir(source_image_dir)) ==  len(os.listdir(source_label_dir))
```

## 2. Rename frames

```python
image_lists = sorted(os.listdir(source_image_dir))
```

```python
image_lists[:3]
```

```python
for i, image_file in enumerate(image_lists):
    convert_name = f"{i:010d}"
    label_file = os.path.splitext(image_file)[0]+".txt"
    # image
    os.rename(os.path.join(source_image_dir,image_file), os.path.join(source_image_dir,convert_name+".jpg"))
    # label
    os.rename(os.path.join(source_label_dir,label_file), os.path.join(source_label_dir,convert_name+".txt"))
```
